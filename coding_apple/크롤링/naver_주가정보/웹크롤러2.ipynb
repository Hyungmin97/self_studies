```python
import requests
from bs4 import BeautifulSoup
```

```python
data = requests.get("https://finance.naver.com/item/sise.naver?code=007070")
```

```python
soup = BeautifulSoup(data.content, 'html.parser')
```

```python
print(soup.find_all('strong', id="_nowVal")[0].text)
```

```python
print(soup.find_all('span', class_="tah")[5].text)
```

```python
#동일업종 등락률
print(soup.find_all('em')[0].text)

soup.select('.f_down em')[0].text
```

```python
#이미지 찾기
image = soup.select('#img_chart_area')[0]
print(image['src'])
```
